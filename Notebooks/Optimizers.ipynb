```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

def func_x(x):
    return (x+5)**2

x = np.linspace(-10,0,30)

plt.plot(x, func_x(x))
plt.show()
```

```python
def func_x_y(x, y):
    return 1 + (0.1 * (y**2 + (2*(x**2))))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = func_x_y(X, Y)

fig = plt.figure(figsize=(14,6))

ax = fig.add_subplot(1, 1, 1, projection='3d')
p = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=matplotlib.cm.coolwarm, linewidth=0, antialiased=False)
cb = fig.colorbar(p, shrink=0.5)
ax.set_zlim3d(0, 15)
```

## **Gradient Descent**
* One variable



```python
x_current = 3
learning_rate = 0.01
precision = 0.000001
delta_x = 1
max_iters = 10000
iters = 0

fx = lambda x: (x+5)**2
dfx = lambda x: 2*(x+5)

x_values_gradient_descent = []

while delta_x > precision and iters < max_iters:
    x_previous = x_current
    x_current = x_current - learning_rate*dfx(x_previous)
    delta_x = abs(x_current - x_previous)
    iters = iters + 1
    if(iters % 50 == 0):
        print(f"Iteration : {iters}, Value : {x_current}")
    
    x_values_gradient_descent.append(x_current)

print(f"The Local minimum occurs at : {x_current} and the value is : {fx(x_current)}")
print(f"Total iterations taken to reach local minima : {iters}")

# plot the gradient descent algorithm
x_values_gradient_descent = np.array(x_values_gradient_descent)
y_values_gradient_descent = fx(x_values_gradient_descent)
plt.plot(x_values_gradient_descent, y_values_gradient_descent)
plt.show()
```

* Multiple Independent variables

```python
x_current = 3
y_current = 40
learning_rate = 0.01
precision = 0.000001
delta_xy = 1
max_iters = 10000
iters = 0

fxy = lambda x,y: (1 + 0.1 * (y**2 + 2 * x**2))
dfxy_dx = lambda x,y: 0.1 * 2 * 2 * x
dfxy_dy = lambda x,y: 0.1 * 2 * y

x_values_gradient_descent = []
y_values_gradient_descent = []

while delta_xy > precision and iters < max_iters:
    x_previous = x_current
    y_previous = y_current
    x_current = x_current - learning_rate * dfxy_dx(x_previous, y_previous)
    y_current = y_current - learning_rate * dfxy_dy(x_previous, y_previous)
    iters = iters + 1
    if iters % 1000 == 0:
        print(f"Iteration : {iters}, value for x : {x_current} and y : {y_current}")

    x_values_gradient_descent.append(x_current)
    y_values_gradient_descent.append(y_current)
    
print(f"The local minima occurs at x : {x_current} and y : {y_current}")

# plot the gradient descent steps
x_values_gradient_descent = np.array(x_values_gradient_descent)
y_values_gradient_descent = np.array(y_values_gradient_descent)
plt.plot(x_values_gradient_descent, y_values_gradient_descent)
```

## **Gradient descent: ADAM (Adaptive Moment Estimate) using Multiple Independent Variables**

```python
x_current = 3
y_current = 40
learning_rate = 0.1
precision = 0.000001
delta_xy = 1
max_iters = 10000
iters = 0

beta_1 = 0.9
beta_2 = 0.999

epsilon = 1e-8

fxy = lambda x,y: (1 + 0.1 * (y**2 + 2 * x**2))
dfxy_dx = lambda x,y: 0.1 * 2 * 2 * x
dfxy_dy = lambda x,y: 0.1 * 2 * y

x_values_gradient_descent = []
y_values_gradient_descent = []

mt_x_previous = 0.0
mt_y_previous = 0.0
vt_x_previous = 0.0
vt_y_previous = 0.0

while delta_xy > precision and iters < max_iters:
    x_previous = x_current
    y_previous = y_current

    gradient_x = dfxy_dx(x_previous, y_previous)
    gradient_y = dfxy_dy(x_previous, y_previous)

    mt_x = beta_1 * mt_x_previous + (1 - beta_1) * gradient_x
    mt_y = beta_1 * mt_y_previous + (1 - beta_1) * gradient_y
    vt_x = beta_2 * vt_x_previous + (1 - beta_2) * gradient_x**2
    vt_y = beta_2 * vt_y_previous + (1 - beta_2) * gradient_y**2

    mt_x_hat = mt_x / (1 - beta_1**(iters + 1))
    mt_y_hat = mt_y / (1 - beta_1**(iters + 1))
    vt_x_hat = vt_x / (1 - beta_2**(iters + 1))
    vt_y_hat = vt_y / (1 - beta_2**(iters + 1))

    x_current = x_current - ((learning_rate * mt_x_hat) / ((vt_x_hat)**0.5 + epsilon))
    y_current = y_current - ((learning_rate * mt_y_hat) / ((vt_y_hat)**0.5 + epsilon))

    delta_xy = ((x_current - x_previous)**2 + (y_current - y_previous)**2)**0.5
    
    iters = iters + 1
    if iters % 1000 == 0:
        print(f"Iteration : {iters}, value of x : {x_current} and y : {y_current}")

    x_values_gradient_descent.append(x_current)
    y_values_gradient_descent.append(y_current)

print(f"The local minima occurs at x : {x_current} and y : {y_current}")

x_values_gradient_descent = np.array(x_values_gradient_descent)
y_values_gradient_descent = np.array(y_values_gradient_descent)
plt.plot(x_values_gradient_descent, y_values_gradient_descent)
plt.show()
```
